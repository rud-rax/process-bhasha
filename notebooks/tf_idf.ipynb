```python
from tf_idf.tf_idf import tf_idf_vectorizer
from imports import Lemmatization
```

```python
documents = pd.read_csv(r"data/Twitter_Data.csv")
```

```python
documents.head()
```

```python
documents.drop(columns= ['category'] , inplace = True)
```

```python
documents = documents.head(25)
```

```python
documents = list(documents['clean_text'])
```

```python
documents
```

```python
documents = [Lemmatization.run('wordnet',text) for text in documents]
```

```python
documents
```

```python
feature_names , tfidf_matrix = tf_idf_vectorizer(documents)
```

```python
# Print the TF-IDF vectors
print("TF-IDF Vectors:")
for i, doc in enumerate(documents):
    print(f"Document {i+1}:")
    for j, term in enumerate(feature_names):
        tfidf_value = tfidf_matrix[i, j]
        if tfidf_value != 0:
            print(f"{term}: {tfidf_value:.3f}")
    print()
```
